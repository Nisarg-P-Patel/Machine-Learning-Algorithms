```python
import numpy,random
from sklearn import neighbors as nb
from sklearn import datasets as dt
from sklearn import metrics as mt
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV

X,Y=dt.load_iris(return_X_y=True)

X_train=X[range(0,150,2),:]
Y_train=Y[range(0,150,2)]

X_test=X[range(1,150,2),:]
Y_test=Y[range(1,150,2)]
```

```python
def print_final(h_size,iteration_size,X_test,Y_test):
    MLP=MLPClassifier(solver='sgd',hidden_layer_sizes=(h_size),max_iter=iteration_size).fit(X_train,Y_train)
    prediction = MLP.predict(X_test)

    print(prediction)
    # print(MLP.predict_proba(X_test))

    print("\nAccuracy with normalize = ",mt.accuracy_score(Y_test,prediction,normalize=True))
    print(mt.classification_report(Y_test,prediction))
    print(mt.confusion_matrix(Y_test,prediction))
def accuracy_final(h_size,iteration_size,X_test,Y_test):
    MLP=MLPClassifier(solver='sgd',hidden_layer_sizes=(h_size),max_iter=iteration_size).fit(X_train,Y_train)

    return mt.accuracy_score(Y_test,MLP.predict(X_test),normalize=True)
```

```python
temp=[]
accu=[]
for x in range(0,15):
    h=random.randrange(80,100,1)
    i=random.randrange(300,500,25)
    temp.append([h,i])
    accu.append(accuracy_final(h,i,X_test,Y_test))

print("Hidden Layer Number :- ",end=" ")
print(temp[accu.index(max(accu))][0])
print("Number of iterations :- ",end=" ")
print(temp[accu.index(max(accu))][1])
print_final(temp[accu.index(max(accu))][0],temp[accu.index(max(accu))][1],X_test,Y_test)
```

```python
param_grid = [
        {
            'activation' : ['identity', 'logistic', 'tanh', 'relu'],
            'solver' : ['lbfgs', 'sgd', 'adam'],
            'hidden_layer_sizes': [
             (81,), (82,),(83,),(84,),(85,),(86,),(87,),(88,),(89,),(90,),
             (91,), (92,),(93,),(94,),(95,),(96,),(97,),(98,),(99,),(100,)
             ],
            'max_iter':[300,325,350,375,400,425,450,475,500]
        }
       ]
MLP_Grid = GridSearchCV(MLPClassifier(), param_grid, cv=3, scoring='accuracy').fit(X_train,Y_train)


print("Best parameters set found on development set:")
print(MLP_Grid.best_params_)
print(MLP_Grid.predict(X_test))
print(mt.accuracy_score(Y_test,MLP_Grid.predict(X_test),normalize=True))
```
