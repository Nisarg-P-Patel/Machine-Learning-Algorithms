```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import metrics as mt

data=pd.read_csv('Train.csv')
X=data['x']
Y=data['y']
X_train=np.ones((699,2))
X_train[:,1]=np.array(X)
Y_train=np.array(Y)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
theta=(0,0)

print(X_train.shape)
m=X_train.shape[0]
n=X_train.shape[1]

niter=100
alpha=0.0001

for i in range(niter):
    update=np.zeros(X_train.shape[1])
    Ynw=np.dot(X_train,theta)
    err=Ynw-Y_train
    for j in range(n):
        update[j] = np.sum(np.dot(err,X_train[:,j]))
    theta=theta-(1/m)*(alpha)*update
print(theta)

PredY=theta[0]+(X_train[:,1]*theta[1])

fig=plt.figure(figsize=(8,8))
plt.scatter(X_train[:,1],Y_train,color="grey")
plt.plot(X_train[:,1],PredY,color="black")
```

```python
data=pd.read_csv('Test.csv')

X=data['x']
Y=data['y']
X_test=np.array(X)
Y_test=np.array(Y)

PredYTest=theta[0]+(X*theta[1])

plt.scatter(X,Y,color="grey")
plt.plot(X,PredYTest,color="blue")
```

```python
print("Absolute_GD:",mt.mean_absolute_error(Y_test,PredYTest))
print("Squared_GD:",mt.mean_squared_error(Y_test,PredYTest))
```
