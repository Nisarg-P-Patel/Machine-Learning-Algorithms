```python
import numpy
import pandas as pd
from sklearn import datasets as dt
from sklearn import metrics as mt
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X,Y=dt.load_iris(return_X_y=True)

X_train=X[range(0,150,2),:]
Y_train=Y[range(0,150,2)]

X_test=X[range(1,150,2),:]
Y_test=Y[range(1,150,2)]
```

```python
model=KMeans(n_clusters=3, init = 'k-means++', max_iter = 350, random_state = 2)
model.fit(X_train)

print(model.labels_)
```

```python
prediction=model.predict(X_test)

print(prediction)
```

```python
print(model.cluster_centers_)
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=prediction,s=50)
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')
```

```python
plt.scatter(X_test[:,0],X_test[:,1],c=Y_test,s=50)
```

```python
print("\nAccuracy = ",mt.accuracy_score(Y_test,prediction))
```
